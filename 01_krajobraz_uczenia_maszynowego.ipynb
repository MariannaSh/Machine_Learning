```python
datapath = Path() / "data" / "LifeSat" 
oecd_bli = pd.read_csv(datapath / "oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

```python
# Wstępnie przetwarzamy dane PKB per capita tak, żeby był uwzględniony wyłącznie 2020 rok i upraszczamy nazwy kolumn
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020 = gdp_per_capita_2020.rename( columns={"GDP per capita, PPP (constant 2017 international $)": "GDP per capita (USD)"})
gdp_per_capita_2020
```

```python
oecd_bli["Indicator"].value_counts()
```

```python
#Dlatego musimy utworzyć tabelę przestawną (pivot table) z interesującymi nas wartościami
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index = "Country", columns = "Indicator", values = "Value")
oecd_bli
```

```python
oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
gdp_per_capita_2020.info()
```

```python
oecd_bli.info()
```

```python
full_country_stats = oecd_bli.merge(gdp_per_capita_2020, left_on = "Country", right_on= "Entity", how = "inner")
full_country_stats
```

```python
full_country_stats = full_country_stats[["Country", "Life satisfaction", "GDP per capita (USD)"]]
full_country_stats = full_country_stats.sort_values(by = "GDP per capita (USD)")
full_country_stats
```

Wyeksportujmy pełny zbiór 'lifesat_full' do pliku

```python
full_country_stats.to_csv(datapath / "lifesat_full.csv", index = False)
```

```python
lifesat = pd.read_csv("data/LifeSat/lifesat.csv")


lifesat.rename(
    columns={"GDP per capita (USD)": "PKB na mieszkanca (USD)"}, inplace=True
)

lifesat.rename(columns={"Life satisfaction": "Satysfakcja z zycia"}, inplace=True)

lifesat
```

```python
lifesat.plot(kind='scatter', grid = True,
             x="PKB na mieszkanca (USD)", y="Satysfakcja z zycia")
plt.show()

```

```python
#zmienne w moidelu

X = lifesat[["PKB na mieszkanca (USD)"]]
y = lifesat[["Satysfakcja z zycia"]]

#wybieram model
model=LinearRegression()

#trenowanie modelu
model.fit(X, y)


#obliczane wspolczynniki
theta_0=model.intercept_[0]
theta_1=model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")
```

```python
#wizualizujemy
lifesat.plot(kind='scatter', grid=True,
             x="PKB na mieszkanca (USD)", y="Satysfakcja z zycia")
plt.plot(X, theta_0+theta_1*X)
plt.text(45000, 6.5, fr"$\theta_0 = {theta_0.round(2)}$", color="blue")
plt.text(45000, 6.3, fr"$\theta_1 = {theta_1 * 1e5:.2f}\cdot10^{{-5}}$", color="blue")
plt.show()
```

```python
cyprus_gdp_per_capita_2020 = (
    gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP per capita (USD)"]]
)
cyprus_gdp_per_capita_2020.rename(
    columns={"GDP per capita (USD)": "PKB na mieszkanca (USD)"}, inplace=True)

cyprus_gdp_per_capita_2020
```

```python

#predykcja sszczęscia na mieszkańca na cyprze
cyprus_predicted_life_satisfaction = model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfaction[0][0]
```

```python
#wybieram model 3 najblizszych sasiadów
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=3)

#trenuje model
model_knn.fit(X, y)

#Uzyskuje przewidywania dla Cypru
model_knn.predict(cyprus_gdp_per_capita_2020)
```
